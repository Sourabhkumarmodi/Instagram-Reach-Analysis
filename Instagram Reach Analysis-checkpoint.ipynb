```python
plt.figure(figsize=(10,8))
plt.title("Distribution of Impression From Home")
sns.distplot(data['From Home'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of Impression from Hashtags")
sns.distplot(data['From Hashtags'])
plt.show()
```

```python
data.columns
```

```python
home=data["From Home"].sum()
hashtags=data["From Hashtags"].sum()
explore=data["From Explore"].sum()
other=data['From Other'].sum()
values=[home,hashtags,explore,other]
labels=['From Home','From Hashtags','From Explore','From Other']
fig=px.pie(data,values=values,names=labels,title="Impressions on Instagram Posts From Various Sources",hole=0.25)
fig.show()
```

```python
text = " ".join(i for i in data.Caption)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.style.use('classic')

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text= " ".join(i for i in data.Hashtags)
stopwords= set(STOPWORDS)
wordcloud=WordCloud(stopwords=stopwords, background_color='red').generate(text)
plt.style.use("classic")
plt.figure(figsize=(12,10))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```
